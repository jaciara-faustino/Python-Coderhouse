```python
from datetime import datetime
def alerta(nivel, base, etapa):
    '''
        Alerta de falha de carregamento de base de dados
    '''
    now = str(datetime.now())

    msg = f"Falha no carregamento da base {base} na etapa {etapa}.\n{now}"


    if nivel == 1:
        title = 'ATENÇÃO: Alerta Baixo'
    elif nivel == 2:
        title = 'ATENÇÃO: Alerta Médio'
    elif nivel  == 3:
        title = 'ATENÇÃO: Alerta Alto'
    else:
        print("Nivel",nivel,"não disponível!")

    notification.notify(
            title=title,
            message=msg,
            app_name='alerta',
            timeout=10
        )

alerta(nivel = 2,
        base = "ESTADOS", 
        etapa = "EXTRACAO")
```

```python
import pandas as pd
import requests
```

```python
url = 'https://brasilapi.com.br/api/ibge/uf/v1'
response = requests.get(url)
```

```python
if response.status_code == 200:
    data_json = response.json()
else:
    print("Erro ao acessar a API")   
data_json
```

```python
estados_full = pd.DataFrame(data_json)
estados_full.head(10)
```

```python
tabela_estado_ibge = estados_full[['nome', 'id']]

pd.DataFrame (tabela_estado_ibge)
```

```python
df_regiao_contagem = estados_full['regiao'].value_counts().reset_index()
df_regiao_contagem.columns = ['regiao', 'quantidade_estados']

pd.DataFrame(df_regiao_contagem)
```
